```python

# Here we are trying to read/ print  the list of carbohydrates and only the carbohydrates

import numpy as np
#import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Here were  defining the path to our csv file that will be leverage through the prcess
path = 'C:\\Users\\user\\CSCI 5015\\data\\cereal.csv'

# Here we are trying to read/ print  the list of carbohydrates and only the carbohydrates

data = np.genfromtxt(path, delimiter=",", skip_header=1)
carbs = data[:,8]

```

```python
# Here we are trying to Create the probability density function which we named"myprofunc"
def myprodfunc(x , mu , sigma):
    numerator = np.exp(-0.5*((x-mu)/sigma)** 2)
    denominator = sigma *np.sqrt(2*np.pi)
    return numerator / denominator
```

```python
# Here we are trying Create the normal function which we named "mynormfunc"
def mynormfunc(mu , sigma, start, stop , step):
    xrange = np.arange( start, stop , step)
    yrange = [myprodfunc(x, mu , sigma) for x in xrange]
    return (xrange, yrange)
```

```python
# Here were trying to Plot the probability density function with a mean of 0 and a standard deviation of 1 from x=-5 to x=5.
(xrange, red) = mynormfunc (0 ,1, -5, 5, 0.01)
plt.plot(xrange, red, "r")
plt.show()
```

```python
''' 
    Here we are trying to determine the min  of carbs data first
'''
np.min(carbs)
print(f" The min of carbs data is: {np.min(carbs)}")
```

```python
''' 
    Here we are trying to determine the max  ofcarbs data first
'''
np.max(carbs)
print(f" The max of carbs data is: {np.max(carbs)}")
```

```python
# Here we are trying to building the Kernel Density Estimation of the data set. and plot it 
x = np.arange(np.min(carbs)-2, np.max(carbs)+2, 0.1)
y = np.zeros(x.shape)
for carb in carbs:
    (_, norm) = mynormfunc(carb, 1, np.min(carbs)-2 , np.max(carbs)+2 , 0.1)
    y = y + norm
```

```python
# Here are trying to Plot the KDE of the carbohydrates data set.  
plt.plot(x ,y , "r")
plt.title("Kernel Density Estimation (KDE) ")
plt.xlabel("Carbohydrate Value")
plt.show()
```
